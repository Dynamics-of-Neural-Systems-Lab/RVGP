```python
import os
import numpy as np
import torch
from modules import LatentLowRankRNN, train
from RNN_helpers import plot_field_noscalings, plot_trajectories
import matplotlib.pyplot as plt

from MARBLE import dynamics
```

Simulate some trajectories of the Van der Pol oscillator

```python
def get_trajectories(mus=[0.5], area = [[-3, -3],[3, 3]], n = 50, t = np.arange(0, 10, 0.1)):
    X0_range = dynamics.initial_conditions(n, len(mus), area)

    x, dx = [], []
    for X0, m in zip(X0_range, mus):
        x_, dx_ = dynamics.simulate_vanderpol(m, X0, t)
        min_len = min([u.shape[0] for u in x_])
        x_ = [u[:min_len,:] for u in x_]
        dx_ = [u[:min_len,:] for u in dx_]
        x_ = [torch.tensor(u, dtype=torch.float32) for u in x_]
        dx_ = [torch.tensor(u, dtype=torch.float32) for u in dx_]
        x.append(torch.stack(x_, axis=0))
        dx.append(torch.stack(dx_, axis=0))
        
    x = torch.cat(x, axis=0)
    dx = torch.cat(dx, axis=0)
    label = torch.rand(dx.shape[0], dx.shape[1], 2)
        
    return x, dx, label
```

```python
x_train, dx_train, label = get_trajectories()
```

```python
noise_std = 5e-2
alpha = 0.2
rank = 2
```

```python
net = LatentLowRankRNN(2, [16,32,512], rank=rank, alpha=alpha, noise_std=noise_std)
train(net, x_train, label, epochs=150, batch_size=32, lr=0.01)
```

```python
x_ = x_train[:,[0],:].to(device=net.m.device)
x, _, _ = net(x_, seq_len=x_train.shape[1])
plt.figure()
plt.plot(x[0].detach().cpu().numpy(), 'g--')
plt.plot(x_train[0],'r')
plt.plot()
```

```python
ax, _ = plot_field_noscalings(net, input=None, xmin=-3, xmax=3, ymin=-3, ymax=3, add_fixed_points=False)

_, _, z = net(x_train[0][0:2].unsqueeze(1).to(device=net.m.device))
plot_trajectories(net, z.detach().cpu().numpy() , ax, n_traj=2, style="-", c="C0")
```
